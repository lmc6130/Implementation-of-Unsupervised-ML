Silhouette Score 
*  用於評估聚類質量的指標
*  綜合考慮了聚類內部的相似度和不同聚類之間的差異性，來衡量聚類的分離程度
*  取值範圍從-1到1
*  接近1時，表示聚類內部的樣本相似度高，不同聚類之間的差異性也很大，聚類的分離程度較好
*  接近0時，表示聚類內部的樣本相似度和不同聚類之間的差異性相近，聚類的分離程度一般
*  接近-1時，表示聚類內部的樣本相似度低，不同聚類之間的差異性也很小，聚類的分離程度較差
*  計算需要考慮每個樣本的聚類內部相似度和與其他聚類的相似度
*  可以用於比較不同聚類算法或不同聚類參數下的聚類效果，選擇最佳的聚類方案

# K-means
https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html

## Iris Example

```python
# 適用於線性可分
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.metrics import normalized_mutual_info_score

kmeans = KMeans(n_clusters=3, n_init=10)
kmeans.fit(X)
kmeans_labels = kmeans.labels_

score = silhouette_score(X, kmeans_labels) # cluster from feature X
score2 = normalized_mutual_info_score(y, kmeans_labels) # ground truth, clustering prediction
print('Silhouette Score:', score)
print('NMI Score:', score2)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 2], c=kmeans_labels)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```

```python
silhouette_avgs = []
n_init=10
ks = range(2, 11)
for k in ks:
    kmeans_fit = KMeans(n_clusters = k, n_init=n_init).fit(X)
    kmeans_labels = kmeans_fit.labels_
    silhouette_avg = silhouette_score(X, kmeans_labels)
    silhouette_avgs.append(silhouette_avg)

# 作圖並印出 k = 2 到 10 的績效
plt.bar(ks, silhouette_avgs)
plt.show()
```

```python
sse = []
n_init = 10
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, n_init=n_init)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)

plt.plot(range(2, 11), sse, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```

# DBSCAN
https://scikit-learn.org/stable/modules/generated/sklearn.cluster.DBSCAN.html

## Iris Example

```python
# 適用於有noise的dataset
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.metrics import normalized_mutual_info_score

dbscan = DBSCAN(eps=0.8, min_samples=5)
dbscan.fit(X)
dbscan_labels = dbscan.labels_

score = silhouette_score(X, dbscan_labels)
score2 = normalized_mutual_info_score(y, dbscan_labels) # ground truth, clustering prediction
print('Silhouette Score:', score)
print('NMI Score:', score2)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 2], c=dbscan_labels)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
import matplotlib.pyplot as plt

# Calculate k-distance for different values of k
k_max = 10
nbrs = NearestNeighbors(n_neighbors=k_max).fit(X)
distances, indices = nbrs.kneighbors(X)
k_distance = np.sort(distances[:, k_max-1], axis=0)

# Plot the elbow curve
plt.plot(range(1, len(k_distance)+1), k_distance, marker='o')
plt.xlabel('k')
plt.ylabel('k-distance')
plt.title('Elbow Method for DBSCAN Clustering')
plt.show()
```
